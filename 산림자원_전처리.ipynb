```python
# 데이터 정제:'활착본수'가 0이거나 null인 행 제거
# '활착본수' 열의 null 값 제거
df_cleaned = df.dropna(subset=['활착본수'])

# '활착본수'가 0인 행 제거
df_cleaned = df_cleaned[df_cleaned['활착본수'] != 0]

print(f"원본 데이터프레임 크기: {df.shape}")
print(f"'활착본수'가 0이거나 null인 행 제거 후 데이터프레임 크기: {df_cleaned.shape}")

display(df_cleaned.head())
```

```python
# 결측치 개수 확인
df_cleaned.isnull().sum()
```

```python
# '임산물종류명' 열의 null 값 제거
print(f"원본 데이터프레임 크기: {df_cleaned.shape}")

df_cleaned = df_cleaned.dropna(subset=['임산물종류명'])

print(f"'임산물종류명'이 null인 행 제거 후 데이터프레임 크기: {df_cleaned.shape}")
```

```python
# 결측치 개수 확인
df_cleaned.isnull().sum()
```

```python
# 컬럼별 데이터 타입 확인
df.dtypes
```

```python
# 활착본수, 본수, 조성면적(제곱미터) 컬럼을 정수형으로 변환
# 현재 데이터에 null 값이 없고, 소수점 이하가 의미 없는 본수/면적 데이터이므로 int64로 변환합니다.
df_cleaned['활착본수'] = df_cleaned['활착본수'].astype('int64')
df_cleaned['본수'] = df_cleaned['본수'].astype('int64')
df_cleaned['조성면적(제곱미터)'] = df_cleaned['조성면적(제곱미터)'].astype('int64')

print("\n데이터 타입 변경 후:")
display(df_cleaned.dtypes)
display(df_cleaned.head())
display(df_cleaned.describe())
```

```python
# '조성면적(제곱미터)'가 0인 행 제거
print(f"원본 데이터프레임 크기: {df_cleaned.shape}")

df_cleaned = df_cleaned[df_cleaned['조성면적(제곱미터)'] != 0]

print(f"'임산물종류명'이 null인 행 제거 후 데이터프레임 크기: {df_cleaned.shape}")
display(df_cleaned.describe())
```

```python
# '계획본수, 사유림산림사업번호, 용역사업계약구분, 산림자원조성관리사업구분'  컬럼 삭제
df_cleaned = df_cleaned.drop(columns=['계획본수', '사유림산림사업번호', '용역사업계약구분', '산림자원조성관리사업구분'])
df_cleaned.head()
```

```python
# 컬럼명 변경
df_cleaned = df_cleaned.rename(columns={
    '활착본수': 'settle_trees',
    '기관명': 'location',
    '필지고유번호(PNU)': 'pnu',
    '임산물종류명': 'tree_species',
    '본수': 'planted_trees',
    '조성면적(제곱미터)': 'planted_area_sqm'
})

df_cleaned.head()
```

```python
df_cleaned[['sido', 'sigungu']] = df_cleaned['location'].str.split(' ', n=1, expand=True)
df_cleaned.head()
```

```python
# location 컬럼 삭제
df_cleaned = df_cleaned.drop(columns=['location'])
df_cleaned.head()
```

```python
# 컬럼 순서 재배열
df_cleaned = df_cleaned[[
    'sido',
    'sigungu',
    'pnu',
    'tree_species',
    'planted_trees',
    'settle_trees',
    'planted_area_sqm'
]]

display(df_cleaned.head())
```
